```python
# Para obtener el tipo de valor en la columna "sexo":
tipo_de_valor = dataset['Sexo'].dtype

# Verificar valores faltantes en la columna "Sexo"
valores_faltantes = dataset['Sexo'].isna().sum()

# Muestra el tipo de valor en la columna "sexo"
print("Tipo de dato en la columna 'Sexo':", tipo_de_valor)

# Mostrar la cantidad de valores faltantes
print("Cantidad de valores faltantes en la columna 'Sexo':", valores_faltantes)
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia de LabelEncoder
label_encoder = LabelEncoder()

# Aplicar Label Encoding a la columna "Region" en tu DataFrame
dataset['Region_encoded'] = label_encoder.fit_transform(dataset['Region'])

# Puedes ver las correspondencias entre etiquetas y valores numéricos
print(dict(zip(dataset['Region'], dataset['Region_encoded'])))
```

```python
# Reemplazar los valores faltantes con la moda (valor más común)
moda = dataset['Region'].mode()[0]
dataset['Region'].fillna(moda, inplace=True)
```

```python
# Verificar valores faltante en la columna "Renta"
valores_faltantes = dataset['Renta'].isna().sum()

# Mostrar la cantidad de valores faltantes
print("Cantidad de valores faltantes en la columna 'Renta':", valores_faltantes)
```

```python

# Imputación por Mediana

# Reemplazamos comas por puntos y convertir a tipo float
dataset['Renta'] = dataset['Renta'].str.replace(',', '.').astype(float)
# Calcula la mediana de la columna 'Renta'
mediana_renta = dataset['Renta'].median()

# Reemplaza los valores nulos en 'Renta' con la mediana
dataset['Renta'].fillna(mediana_renta, inplace=True)
```

```python
# Verificar valores imputados en la columna "Renta"
valores_faltantes = dataset['Renta'].isna().sum()

# Mostrar la cantidad de valores faltantes
print("Cantidad de valores faltantes en la columna 'Renta':", valores_faltantes)
```
